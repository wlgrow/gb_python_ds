## Задание 2

```python
t = np.linspace(0, 10, num=51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color="green")
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

## Задание 3

```python
x = np.linspace(-3, 3, num = 51)
x
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title("График y2")

ax3.plot(x, y3)
ax3.set_title("График y3")

ax4.plot(x, y4)
ax4.set_title("График y4")
```

## Задание 4

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
credit_card = pd.read_csv("creditcard.csv", sep=",")
```

```python
c = credit_card['Class'].value_counts()
c
```

```python
plt.style.use("fivethirtyeight")
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
c.plot(kind="bar")
plt.show()
```

```python
c.plot(kind="bar", logy=True)
plt.show()
```

```python
v11 = credit_card[credit_card['Class'] == 1]['V1']
v10 = credit_card[credit_card['Class'] == 0]['V1']
```

```python
plt.hist(v11, density=True, bins=20, alpha=0.5, label="Class 1", color="red")
plt.hist(v10, density=True, bins=20, alpha=0.5, label="Class 0", color="grey")

plt.xlabel("V1")

plt.legend()
plt.show()
```
